```python
import pandas as pd
from datetime import datetime

def collect_expenses():
    expenses=[]
    while True:
        while True:
         date=input(" enter date in YY/MM/DD order")
         try:
          date = datetime.strptime(date, "%Y/%m/%d")
          break  
         except ValueError:
          print("Invalid date format. Please use YY/MM/DD format.")
        category=input(" Enter The category of the expense, such as Food or Travel")
        amount=int(input("enter the amount spent"))
        description=input("A brief description of the expense")
        expense={'date':date,'category':category,'amount':amount,'description':description}
        expenses.append(expense)
        return pd.DataFrame(expenses)
    
def input_budget():
         budget=int(input("enter your monthly budget"))
         return budget
    
def view_expenses(expenses):
        if expenses.empty:
            print("no response recorded")
            return
        for index ,expense in expenses.iterrows():
          print(f"{expense.get('date')} | {expense.get('category')} | ${int(expense.get('amount'))} | {expense.get('description')}")
 
def calculate_expenses(expenses,budget):
         total_expenses = expenses['amount'].sum()
         if total_expenses>budget:
             print(f"you have crossed your monthly expenses{total_expenses}")
         else:
             savings = budget-total_expenses
             print(f"your remaining savings :{savings }")


def save_expenses_to_csv(expenses, filename="expenses.csv"):
               expenses.to_csv(filename, index=False)
               print(f"Expenses saved to {filename}")

def read_csv(expenses,filename="expenses.csv"):
        try:
                          existing_data = pd.read_csv(filename)
      
                          updated_expenses = pd.concat([existing_data, expenses], ignore_index=True)
                          print(f"Expenses read from {filename}")
                          return updated_expenses
        except FileNotFoundError:
                      print(" sorry file not found")
                      return pd.DataFrame(columns=['date', 'category', 'amount', 'description'])
                   
def display_menu():
         print("choose from the given menu")
         print("1.collect expenses")
         print("2.view expenses")
         print("3.track budget")
         print("4.save expenses")
         print("5.exit")

def main():
         expenses=pd.DataFrame(columns=['date','category','amount','description'])
         budget=0
         while True:
                display_menu()
                choice=input("select a option")
                if choice=='1':
                          new_expenses=collect_expenses()
                          expenses=pd.concat([expenses,new_expenses],ignore_index=True)
                elif choice=='2':
                        view_expenses(expenses)
                elif choice=='3':
                         if budget==0:
                                 budget=input_budget()
                                 calculate_expenses(expenses,budget)
                elif choice=='4':
                          combined_expenses = read_csv(expenses)  
                          save_expenses_to_csv(combined_expenses)
                elif choice=='5':
                        print(" exiting the program")
                        break
                else:
                         print("invalid option ,choose again")
if __name__ == '__main__':
 main()


        
```
